```python
# partition raw data into appropriate dataframes
# we start by filtering-out any extraneous jobs
# each subsequent dataset will check against repeat data from the previous dataset 

# other jobs

data_other_jobs = pd.Series([True if (('Programmer' in job_name)
                                    | ('programmer' in job_name)
                                    | ('Consultant' in job_name)
                                    | ('consultant' in job_name)
                                    | ('Manager' in job_name)
                                    | ('manager' in job_name)
                                    | ('PM' in job_name)
                                    | ('Owner' in job_name)
                                    | ('owner' in job_name)
                                    | ('Operations' in job_name)
                                    | ('Director' in job_name)
                                    | ('director' in job_name)
                                    | ('Instructor' in job_name)
                                    | ('instructor' in job_name)
                                    | ('Educator' in job_name)
                                    | ('educator' in job_name)
                                    | ('School' in job_name)
                                    | ('school' in job_name)
                                    | ('Research' in job_name)
                                    | ('research' in job_name)
                                    | ('Fellow' in job_name)
                                    | ('Statistician' in job_name)
                                    | ('statistician' in job_name)
                                    | ('INFORMATICIAN' in job_name))
                             
                             else False for job_name in jobs_raw.job_name])

data_other_df = jobs_raw.loc[data_other_jobs]

# data scientist

data_sci_jobs = pd.Series([True if ((('Scientist' in job_name)
                                    | ('scientist' in job_name)
                                    | ('SCIENTIST' in job_name)
                                    | ('Sciences' in job_name)
                                    | ('AVP' in job_name))
                                    & (job_name not in list(data_other_df.job_name)))
                           
                           else False for job_name in jobs_raw.job_name])

data_sci_df = jobs_raw.loc[data_sci_jobs]

# data engineer

data_eng_jobs = pd.Series([True if ((('Engineer' in job_name)
                                     | ('engineer' in job_name)
                                     | ('ENGINEER' in job_name)
                                     | ('ML' in job_name)
                                     | ('AI' in job_name))
                                     & (job_name not in list(data_other_df.job_name))
                                     & (job_name not in list(data_sci_df.job_name)))
                           
                           else False for job_name in jobs_raw.job_name])

data_eng_df = jobs_raw.loc[data_eng_jobs]

# data analyst

data_anl_jobs = pd.Series([True if ((('Analyst' in job_name)
                                    | ('analyst' in job_name)
                                    | ('ANALYST' in job_name)
                                    | ('Data Specialist' in job_name)
                                    | ('data specialist' in job_name)
                                    | ('Analytics' in job_name)
                                    | ('analytics' in job_name)
                                    | ('Insight' in job_name)
                                    | ('insight' in job_name))
                                    & (job_name not in list(data_other_df.job_name)) 
                                    & (job_name not in list(data_sci_df.job_name))
                                    & (job_name not in list(data_eng_df.job_name)))
                           
                           else False for job_name in jobs_raw.job_name])

data_anl_df = jobs_raw.loc[data_anl_jobs]
```

We will approach our problem with the *bag of words* format

```python
# example

data_sci_new.job_duties[11]
```

after combing through the data, below are the expressions we need to remove with regex:

- punctuation
- \n
- parenthesis

```python
# we will start by defining a function that does this for us

def clean_text(text):
    '''Make text lowercase, remove punctuation, remove apostrophes and remove \n'''
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('[’‘]', '', text)
    text = re.sub('\n', ' ', text)
    return text

clean = lambda x: clean_text(x)
```

```python
# then we drop the null values from all fields of interest

ds_duties = data_sci_new.job_duties.dropna().reset_index(drop = True)

ds_basic_reqs = data_sci_new.job_basic_reqs.dropna().reset_index(drop = True)

ds_desired_reqs = data_sci_new.job_desired_reqs.dropna().reset_index(drop = True)

ds_description = data_sci_new.job_description.dropna().reset_index(drop = True)
```

```python
# then apply cleaning

ds_duties_clean = pd.DataFrame(ds_duties.apply(clean))

ds_basic_reqs_clean = pd.DataFrame(ds_basic_reqs.apply(clean))

ds_desired_reqs_clean = pd.DataFrame(ds_desired_reqs.apply(clean))

ds_description_clean = pd.DataFrame(ds_description.apply(clean))
```

```python
# now create our document term matrix (DTM)

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# create count vectorizer

cv = CountVectorizer(stop_words = "english", max_df = 0.5) # set it 0.5 so anything too common is not considered
```

Let's limit ourselves to the basic requirements for now

```python
# basic reqs

basic_reqs_cv = cv.fit_transform(ds_basic_reqs_clean.job_basic_reqs)
basic_reqs_dtm = pd.DataFrame(basic_reqs_cv.toarray(), columns = cv.get_feature_names())
basic_reqs_dtm.index = ds_basic_reqs_clean.index
```

```python
# everything else

'''
duties_cv = cv.fit_transform(ds_duties_clean.job_duties)
duties_dtm = pd.DataFrame(duties_cv.toarray(), columns = cv.get_feature_names())
duties_dtm.index = ds_duties_clean.index

desired_reqs_cv = cv.fit_transform(ds_desired_reqs_clean.job_desired_reqs)
desired_reqs_dtm = pd.DataFrame(desired_reqs_cv.toarray(), columns = cv.get_feature_names())
desired_reqs_dtm.index = ds_desired_reqs_clean.index

description_cv = cv.fit_transform(ds_description_clean.job_description)
description_dtm = pd.DataFrame(description_cv.toarray(), columns = cv.get_feature_names())
description_dtm.index = ds_description_clean.index
'''
```

```python
# frequency

basic_reqs_top = basic_reqs_dtm.sum().sort_values(ascending = False)

for word in basic_reqs_top.index[0:10]: print(word, " : ", basic_reqs_top.loc[word])
```

Not bad! But lets's attempt to tokenize by bigram

```python
cv2 = CountVectorizer(stop_words = "english", ngram_range = (1,2), max_df = 0.5)

basic_reqs_cv2 = cv2.fit_transform(ds_basic_reqs_clean.job_basic_reqs)
basic_reqs_dtm2 = pd.DataFrame(basic_reqs_cv2.toarray(), columns = cv2.get_feature_names())
basic_reqs_dtm2.index = ds_basic_reqs_clean.index

basic_reqs_top2 = basic_reqs_dtm2.sum().sort_values(ascending = False)

for word in basic_reqs_top2.index[0:10]: print(word, " : ", basic_reqs_top2.loc[word])
```

Very interesting! We can pull a couple key insights from this:

- machine learning is a frequent basic requirement
- python appears to be the most-sought programming language
- applicants should have a degree at the basic level

Let's use a word cloud to visualize the most used terms for each of the roles: Data Scientist, Data Engineer, Data Analyst

```python
from wordcloud import WordCloud
from sklearn.feature_extraction import text
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# designate some other stop words to glean more insight
# note: all I did here was look at the word clouds below and updated the list with extraneous words

other_stop_words = ['data', 'experience', 'work', 'familiarity', 'year', 'years', 'field', 'skills', 'science', 'strong',
                   'ability', 'business', 'including', 'tool', 'tools', 'use', 'related', 'preferred', 'andor', 'environment',
                    'youll', 'role', 'know', 'using', 'paid', 'information']

stop_words = text.ENGLISH_STOP_WORDS.union(other_stop_words)
```

```python
wc = WordCloud(
    max_font_size = 300,
    stopwords = stop_words,
    background_color = 'white',
    colormap = 'Dark2',
    random_state = 2)
```

### Data Scientist

```python
scraped = ['Basic Requirements', 'Desired Requirements', 'Duties', 'Description']

ds_datasets = [ds_basic_reqs_clean, ds_desired_reqs_clean, ds_duties_clean, ds_description_clean]

plt.rcParams['figure.figsize'] = [20, 10]

for index, dataset in enumerate(ds_datasets):

    wc.generate(dataset.to_string())
    
    plt.subplot(2,2,index+1)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title(scraped[index])

plt.show()
```
